```python
# Import Library
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
# List kosong untuk menampung hasil dari scraping data
data = []

# Membuat perulangan secara otomatis untuk mengambil judul artikel dan link dari halaman 1 - halaman 110
for page in range(1, 110):
    if page == 1:
        url = "https://www.detik.com/search/searchall?query=vaksin&siteid=2&sortby=time&page=1"
    else:
        url = "https://www.detik.com/search/searchall?query=vaksin&siteid=2&sortby=time&page=" + str(page)
    
    # Request URL
    req = requests.get(url)
    
    # Merubah format dari yang semula HTML menjadi BeautifulSoup 
    soup = BeautifulSoup(req.text, "html.parser")
    
    # Memanggil seluruh artikel yang berada di halaman luar
    column = soup.findAll("article")
    
    for scraping in column:
        text = scraping.find("h2", "title").text.strip().split("\n")
        kategori = scraping.find("span", "category")
        link = scraping.find("a")["href"]
        print(text, kategori, link)
                
        data.append([text, kategori, link]) # Memasukkan data scraping ke dalam list data
```

```python
# Memasukkan hasil scrap ke dalam DataFrame
df = pd.DataFrame(data)
df.head()
```

```python
# Mengganti nama kolom
df.rename(columns={0: "Text", 1:"Kategori", 2:"URL"}, inplace=True)
df.tail()
```
