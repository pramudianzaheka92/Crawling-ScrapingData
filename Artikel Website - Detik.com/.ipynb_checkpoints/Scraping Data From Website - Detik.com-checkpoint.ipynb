```python
# Import Library
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

### Halaman 1

```python
# Copy URL Website
key = "vaksin"
url = "https://www.detik.com/search/searchall?query={}&siteid=2&sortby=time&page=1".format(key)
```

```python
# Request URL
req = requests.get(url)

# Mercubah format html ke dalam beautiful soup
soup = BeautifulSoup(req.text, "html.parser")

# Mengambil seluruh judul artikel paling terluar
content = soup.findAll("article")

data = [] # List kosong untuk menampung hasil scraping

for article in content:
    text = article.find("h2", "title").text.strip().split("\n")
    kategori = article.find("span", "category")
    link = article.find("a")["href"]
    data.append([text, kategori, link])
```

```python
df1 = pd.DataFrame(data)
df1.head()
```

```python
# Mengganti nama kolom
df1.rename(columns={0: "Text", 1:"Kategori", 2:"URL"}, inplace=True)
df1.tail()
```

```python
df1.info()
```

```python
# Request URL
req = requests.get(url)

# Mercubah format html ke dalam beautiful soup
soup = BeautifulSoup(req.text, "html.parser")

# Mengambil seluruh judul artikel paling terluar
content = soup.findAll("article")

data = [] # List kosong untuk menampung hasil scraping

for article in content:
    text = article.find("h2", "title").text.strip().split("\n")
    kategori = article.find("span", "category")
    link = article.find("a")["href"]
    data.append([text, kategori, link])
```

```python
# Memasukkan hasil scrap ke dalam DataFrame
df26 = pd.DataFrame(data)
df26.head()
```

```python
# Mengganti nama kolom
df26.rename(columns={0:"Text", 1:"Kategori", 2:"URL"})
df26.tail()
```
