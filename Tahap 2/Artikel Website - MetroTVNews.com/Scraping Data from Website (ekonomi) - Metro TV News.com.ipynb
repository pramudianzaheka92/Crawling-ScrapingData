```python
# Import library
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

```python
# List kosong
data = []

# Membuat perulangan secara otomatis untuk mengambil judul artikel dan link dari halaman 1 - halaman 45
for page in range(1, 45):
    if page == 1:
        url = "https://www.metrotvnews.com/search?query=ekonomi"
    else:
        url = "https://www.metrotvnews.com/search/ekonomi/"+ str(page)
      
    # Request URL Website
    page = requests.get(url)
    
    # Parsing HTML ke dalam BeautifulSoup
    soup = bs(page.text, "html.parser")
    
    # Memanggil judul artikel
    article = soup.findAll("div", "item")
    
    # Mengambil judul artikel dan link
    for scraping in article:
        text = scraping.find("h3").text.strip().split("\n")
        link = scraping.find("a")["href"]
        print(text)
                
        data.append([text, link]) # Memasukkan data scraping ke dalam list data
```

```python
# Memasukkan hasil scraping ke dalam DataFrame
df = pd.DataFrame(data)
df.head()
```

```python
# Mengganti nama kolom
df.rename(columns={0: "Judul Berita", 1:"URL"}, inplace=True)
df.tail()
```

```python
# Menjadikan ke dalam format .csv
df.to_csv("Data Scraping (Metro TV News) - Ekonomi.csv", index=False)
```
