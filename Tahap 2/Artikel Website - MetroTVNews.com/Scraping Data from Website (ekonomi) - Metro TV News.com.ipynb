```python
# Import library
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

### Halaman 1

```python
# Copy URL Website
keyword = "ekonomi"
url = "https://www.metrotvnews.com/search?query={}".format(keyword)
```

```python
# Request URL Website
req = requests.get(url)

# Parsing HTML ke dalam BeautifulSoup
soup = bs(req.text, "html.parser")

# Memanggil judul artikel
article = soup.findAll("div", "item")

# List kosong
data = []

for scraping in article:
    text = scraping.find("h3").text.strip().split("\n")
    link = scraping.find("a")["href"]
    
    data.append([text, link]) # Memasukkan data scraping ke dalam list data
```

```python
# Memasukkan hasil scraping ke dalam DataFrame
df1 = pd.DataFrame(data)
df1.head()
```

```python
# Mengganti nama kolom
df1.rename(columns={0: "Judul Berita", 1:"URL"}, inplace=True)
df1.tail()
```
