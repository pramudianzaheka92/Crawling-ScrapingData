```python
# Import library
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# List kosong untuk menampung hasil dari scraping
data = []

# Membuat perulangan secara otomatis untuk mengambil judul artikel dan link dari halaman 1 - halaman 50
for page in range(1, 50):
    if page == 1:
        url = "https://www.metrotvnews.com/search?query=politik"
    else:
        url = "https://www.metrotvnews.com/search/politik/"+ str(page)
      
    # Request URL Website
    page = requests.get(url)
    
    # Parsing format HTML ke dalam BeautifulSoup
    soup = BeautifulSoup(page.text, "html.parser")
    
    # Memanggil judul artikel yang berada diluar
    article = soup.findAll("div", "item")
    
    # Mengambil judul artikel dan link
    for scraping in article:
        text = scraping.find("h3").text.strip().split("\n")
        link = scraping.find("a")["href"]
        print(text)
                
        data.append([text, link]) # Memasukkan data scraping ke dalam list data
```

```python
# Memasukkan hasil scraping ke dalam DataFrame
data_scraping = pd.DataFrame(data)
data_scraping.head()
```

```python
# Mengganti nama kolom di DataFrame "data_scraping"
data_scraping.rename(columns={0: "Judul Berita", 1:"URL"}, inplace=True)
data_scraping.head()
```

```python
# Menjadikan format DataFrame ke dalam format .csv
data_scraping.to_csv("Data Scraping (Metro TV News) - Ekonomi.csv", index=False)
```
